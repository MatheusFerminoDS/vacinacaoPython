## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1- Calcule o percentual de acerto dos alunos por assunto.
df_mat['Percentual_acertos'] = df_mat['Qt_acertos'] / 24 * 100
print("Matemática: ", df_mat['Percentual_acertos'])

df_por['Percentual_acertos'] = df_por['Qt_acertos'] / 18 * 100
print("Português: ", df_por['Percentual_acertos'])

df_geo['Percentual_acertos'] = df_geo['Qt_acertos'] / 8 * 100
print("Geografia: ", df_geo['Percentual_acertos'])

df_ing['Percentual_acertos'] = df_ing['Qt_acertos'] / 8 * 100
print("Inglês: ", df_ing['Percentual_acertos'])

df_his['Percentual_acertos'] = df_his['Qt_acertos'] / 8 * 100
print("História: ", df_his['Percentual_acertos'])

df_fis['Percentual_acertos'] = df_fis['Qt_acertos'] / 12 * 100
print("Física: ", df_fis['Percentual_acertos'])

df_qui['Percentual_acertos'] = df_qui['Qt_acertos'] / 12 * 100
print("Química: ", df_qui['Percentual_acertos'])
````

````python
# 2- Calcule o total de acertos de cada aluno.
df_total = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']
print("\n2- O total de acertos de cada aluno é: ", df_total)
````

````python
# 3- Calcule o porcentual geral de cada aluno.
df_percentual = df_total / 90 * 100
print("\n3- O percentual geral de cada aluno é: ", df_percentual)
````

````python
# 4- Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?
num_students_above_45 = len(df_total[df_total > 45])
print("\n4- O número de alunos que obtiveram nota maior que 45 é: ", num_students_above_45)
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# 1- Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo DataFrame com os alunos desqualificados por este critério.
df_disqualified = pd.concat([df_mat, df_fis, df_qui], axis=1)
df_disqualified.columns = ['Matemática', 'Física', 'Química']
df_disqualified = df_disqualified[(df_disqualified['Matemática'] == 0) | (df_disqualified['Física'] == 0) | (df_disqualified['Química'] == 0)]
print("1- DataFrame com os alunos desqualificados:\\n", df_disqualified)

# 2- Quantos são esses alunos?
num_disqualified = len(df_disqualified)
print("\\n2- Número de alunos desqualificados: ", num_disqualified)

# 3- Qual a média desses alunos em história e geografia?
df_hist_geo = pd.concat([df_his, df_geo], axis=1)
df_hist_geo.columns = ['História', 'Geografia']
df_hist_geo = df_hist_geo.loc[df_disqualified.index]
mean_hist = df_hist_geo['História'].mean()
mean_geo = df_hist_geo['Geografia'].mean()
print("\\n3- Média dos alunos desqualificados em história: ", mean_hist)
print("Média dos alunos desqualificados em geografia: ", mean_geo)

# 4- Monte um DataFrame com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.
df_passed = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)
df_passed.columns = ['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química']
df_passed = df_passed.drop(df_disqualified.index)
df_passed = df_passed[df_passed.sum(axis=1) > 45]
print("\\n4- DataFrame com os alunos que passaram para a segunda fase:\\n", df_passed)
````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vcinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# 1- Crie uma cópia do DataFrame original, contendo somente os registros de vacinas realizadas a menores de 18 anos.
df_under_18 = df[df['paciente_idade'] < 18].copy()
print("DataFrame com registros de vacinas realizadas a menores de 18 anos:\n", df_under_18)
````

````python
# 2- Crie uma lista das colunas desse DataFrame com o atributo de DataFrame .columns()
columns = df_under_18.columns.tolist()
print("\nLista das colunas do DataFrame:\n", columns)
````

````python
# 3- Nesse DataFrame faça uma contagem do campo vacina_categoria_nome.
df_under_18['count'] = df_under_18.groupby('vacina_categoria_nome')['vacina_categoria_nome'].transform('count')
print("\nDataFrame após contagem do campo vacina_categoria_nome:\n", df_under_18)
````

````python
# 4- A partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.
establishment_columns = [col for col in columns if 'estabelecimento' in col]
print("\nLista de colunas referentes ao estabelecimento:\n", establishment_columns)
````

````python
# 5- Usando o método .loc, selecione somente essas variáveis
df_under_18 = df_under_18.loc[:, establishment_columns + ['count']]
print("\nDataFrame após selecionar somente as variáveis referentes ao estabelecimento:\n", df_under_18)
````

````python
# 6- Aplique o método .drop_duplicates e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento
df_under_18 = df_under_18.drop_duplicates()
print("\nDataFrame após aplicar o método .drop_duplicates:\n", df_under_18)
````
